```python
import numpy as np
import seaborn as sns
import scipy
```

```python
dist_dist = scipy.stats.skewnorm.rvs(2, 30, scale=25, size=3000)
```

```python
dist_dist = dist_dist[dist_dist > 0]
```

```python
sns.distplot(dist_dist)
```

```python
def get_kon(dist, mu=23, sigma=5):
    if dist > 35 or dist < 5:
        return 0
    return (1/(sigma*np.sqrt(2*np.pi)) * np.exp(-0.5*((dist-mu)/sigma)**2))
```

```python
kon_dist = np.array([get_kon(dist) for dist in dist_dist])
```

```python
get_kon(23)
```

```python
sns.distplot(-np.log(kon_dist[kon_dist != 0]), kde=False)
```

```python
-np.log(kon_dist[kon_dist > 0].min())
```
